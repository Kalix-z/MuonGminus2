Creating a 4x4 Grid to show time (ns), r (mm), y (mm), and MoM (Gev)

```python
# the size of each graph inside the larger set of graphs
xSize = 17
ySize = 17
# Creates a 4x4 plot array
fig, axs = plt.subplots(4, 4, figsize=(xSize, ySize))
plt.subplots_adjust(hspace=0.42)  # Adjust vertical and horizontal spacing
#plt.tight_layout()
# create an instance of CosyPlot with figure and axes given
plot_main = cosyplot.CosyPlot(fig, axs)
# fills all of the plots with empty plots
for i in range(4):
    for j in range(4):
        plot_main.plot_empty(i, j)

#plot specific plots with specified bandwidth and resolution
#plot_main.plot(cosyplot.COSYInputType.TIME, cosyplot.COSYInputType.TIME, bandwidth=0.005, resolution=1000j, showPoints=False)  
plot_main.plot(cosyplot.COSYInputType.TIME, cosyplot.COSYInputType.MoM, xBand=0.03, yBand=0.75, resolution=100, showPoints=True)  
 
```
